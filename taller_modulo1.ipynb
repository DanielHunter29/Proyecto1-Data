### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
promedio = 0
suma = 0
contador = 0
maximo = None
minimo= None
# Promedio
for i in esperanzas_de_vida:
    contador += 1
    suma += i
# Máximo
    if maximo is None or i > maximo:
        maximo = i

# Mínimo
    if minimo is None or i < minimo:
        minimo = i
promedio = suma / contador
print(promedio)
print(maximo)
print(minimo)


````

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(años):
    if años >= 75:
        return "Alta"
    elif años >= 60 and años <= 74.9:
        return "Media"
    elif años < 60:
        return "Baja"
#clasificar_esperanza_vida(0)
#clasificar_esperanza_vida(20)
#clasificar_esperanza_vida(50)
clasificar_esperanza_vida(75)
#clasificar_esperanza_vida(90)

````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
#df.info()
#df.describe()
media_esperanza = df["esperanza_vida"].mean()
df["esperanza_vida"] = df["esperanza_vida"].fillna(media_esperanza)

media_gasto = df["gasto_salud_usd"].mean()
df["gasto_salud_usd"] = df["gasto_salud_usd"].fillna(media_gasto)

media_mortalidad = df["mortalidad_infantil"].mean()
df["mortalidad_infantil"] = df["mortalidad_infantil"].fillna(media_mortalidad)

media_medicos = df["medicos_por_1000"].mean()
df["medicos_por_1000"] = df["medicos_por_1000"].fillna(media_medicos)

media_obesidad = df["obesidad_pct"].mean()
df["obesidad_pct"] = df["obesidad_pct"].fillna(media_obesidad)

media_diabetes = df["diabetes_pct"].mean()
df["diabetes_pct"] = df["diabetes_pct"].fillna(media_diabetes)

media_tabaquismo = df["tabaquismo_pct"].mean()
df["tabaquismo_pct"] = df["tabaquismo_pct"].fillna(media_tabaquismo)

media_camas = df["camas_hospital_por_1000"].mean()
df["camas_hospital_por_1000"] = df["camas_hospital_por_1000"].fillna(media_camas)

media_vacunacion = df["vacunacion_pct"].mean()
df["vacunacion_pct"] = df["vacunacion_pct"].fillna(media_vacunacion)

df.info()
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: Usamos la media ya que esos datos son importantes y no se deberian eliminar, con esos datos puedo sacar varios datos relevantes.
